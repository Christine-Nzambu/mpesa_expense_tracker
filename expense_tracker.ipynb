**M-pesa Expenses Tracker - July 2025
This project analyzes my real M-Pesa spending using Python (pandas and matplotlib). I categorized expenses, visualized spending trends, and identified my top spending categories. The analysis helped me realize I spend a lot on *furniture*, and I can adjust my budget moving forward.**

```python
# loading the dataset
import pandas as pd
df = pd.read_csv("MY JULY EXPENSES.csv")
```

```python
# printing my dashboard
print("DASHBOARD")
print("Total Expenses: KES", df["Amount"].sum())
print("Average daily spend: ", df.groupby("Date")["Amount"].sum().mean())
print("Highest Expense Category: ", df.groupby("Category")["Amount"].sum().idxmax())
```

```python
# group expenses by category and sum
df.groupby('Category')['Amount'].sum().sort_values(ascending=False)
```

Above we have the total spending per category.

```python
# finding missing or null entries
df.isnull()
```

Checking for null values and we can see that the last row has null entries.

```python

# removing null entries
df.dropna()
```

Removing null values to clean the data.

```python
# looking for duplicates in the data
df.duplicated()
```

Checking for any duplicates in the data. And there are no duplicates.

```python

# plotting a bar chart to visualize total spending per category
import matplotlib.pyplot as plt

df.groupby('Category')['Amount'].sum().plot(kind = 'bar')
plt.title('Total spending per category', c = 'r')
plt.xlabel('Category', c = 'c')
plt.ylabel('Amount', c = 'c')
plt.xticks(rotation = 45)
plt.show()
```

The bar chart above is a visualization of the total amount spend in each category.

The 'furniture' category has the highest spending.

```python

# grouping by total amount spend per day
df.groupby('Date')['Amount'].sum().sort_values(ascending= False)
```

We can see the total amount spend each day.

```python

# plotting a bar chart to show the total amount spend per day
import matplotlib.pyplot as plt
import pandas as pd

df.groupby('Date')['Amount'].sum().plot(kind = 'bar')
plt.title('Total spending per Day', c = 'g')
plt.xlabel('Date', c = 'm')
plt.ylabel('Amount', c = 'm')

plt.show()
```

The bar chart above is a visualization of our data showing the total amount spend per day.

**July 4th** has the highest spending and **July 12th** has the least spending.
